# Metrics Description

explained by copilot:

- `s`: This represents the elapsed time in seconds.
- `h:m:s`: This represents the elapsed time in hours, minutes, and seconds.
- `max_rss`: This is the maximum Resident Set Size.
            It represents the portion of the process's memory that is held in RAM.
- `max_vms`: This is the maximum Virtual Memory Size. 
            It represents the total virtual memory used by the process.
- `max_uss`: This is the maximum Unique Set Size. 
            It represents the portion of the process's memory that is held in RAM 
            and is unique to the process.
- `max_pss`: This is the maximum Proportional Set Size. 
            It represents the portion of the process's memory that is held in RAM 
            and is shared among other processes, proportionally divided.
- `io_in`: This represents the amount of data read by the process from disk
- `io_out`: This represents the amount of data written by the process to disk 
- `mean_load`: This represents the average system load while the process was running.
- `cpu_time`: This represents the total CPU time consumed by the process, in seconds.

```python
benchmarkpath = Path("benchmarks/113_HM10395/")

# list files in benchmark path
files = list(benchmarkpath.glob("*.txt"))
for f in files:
    print(f)
```

```python
column = 0
plt.figure(figsize=(10, 10))
for file in files:
    # Read the data from file
    df = pd.read_csv(file, sep='\t')
    # Plot box-whisker for the first column
    plt.boxplot(df.iloc[:, column], positions=[files.index(file)], widths=0.6, patch_artist=True, showmeans=True)
    
plt.xticks(range(len(files)), [file.stem for file in files], rotation=80)
plt.title('Time Elapsed (s)')
plt.xlabel('Files')
plt.ylabel('seconds')
plt.grid(True)
plt.tight_layout()
plt.show()
```
